```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
```

## 1. Load Data

```python
df = pd.read_csv('cleaned.csv')
```

```python
y = df['Tip Percentage']
percentage_float = [float(y[i][:-1]) for i in range(len(y))]
df['Tip Perc'] = percentage_float
y = df['Tip Perc']
```

```python
df['Tip Perc'].hist(bins=40,color='skyblue')
```

This is the distribution of average tip percentage in different zipcode area in New York City.
We can see from the plot most of tip percentage are around 15%-20%.
However there are some outliers, like the one below 5% and the one more than 35%.
So, we need to first detect outlier and rule them out of our analysis.

We assume the distribution of tip percentage is satisfied normal (or t) distribution. And those out of the 95% confidence interval will be regarded as outliers. We calculate the 95% confidence interval in this way: (median-3*std, median+3*std). Please note generally people choose mean value as center but we choose the median, since there are only 200 instances in our sample and median value should be more robust.

```python
def is_outlier(data, point):
    med = data.median()
    std = data.std()
    lower_limit = med - 3*std
    upper_limit = med + 3*std
    if point < upper_limit and point > lower_limit:
        outlier = False
    else:
        outlier = True
    return outlier
```

```python
data = df['Tip Perc']
```

```python
for i in range(len(data)):
    if is_outlier(data, data[i]):
        print i,data[i]
```

```python
df['Is_Outlier'] = [is_outlier(df['Tip Perc'], df['Tip Perc'][i]) for i in range(len(df))]
```

```python
df_cleaned = df[df['Is_Outlier'] == False]
```

## 2. Roughly OLS

```python
X_labels = list(df_cleaned.columns)[2:-2]
X = df_cleaned[X_labels]
X = sm.add_constant(X)
y = df_cleaned['Tip Perc']
```

```python
result = sm.OLS(y, X, cov_type='robust').fit()
result.summary()
```

We first roughly tried the OLS regression on tip percentage and census data. We chose 10 features based on our common sense. R^2 for this regression is 0.363 (not bad in emprical study). However, if we consider the effect of each factor, then we found noun of them is statistically significant at 95% confidence level (3 of them, renter occupied rate, male rate and average commute to work are significant at 90% confidence level though). This may be an alert for multicollinearity problem. For example, the median household income and Gini index and higher education rate may be correlated. Therefore, we need to separately consider univariable regression for each possible factor.
