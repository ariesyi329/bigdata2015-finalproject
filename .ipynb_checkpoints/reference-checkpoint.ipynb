```python
import argparse,csv,sys,os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from datetime import date,datetime
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.sandbox.regression.predstd import wls_prediction_std


#filename = '2002.csv'
filename = 'CountyCombined.csv'

df = pd.read_csv(filename)
df = df.fillna(method='pad')
```

```python
#df.ix[:50]
#df['Polution'] = df['PercapitaPolution']
df = df.loc[:,:'Density']
df = df[(df['Population'] > 100000)]
df = df[(df['Density'] > 1000)]
df['Density'].describe()
```

```python
df['Population']  = np.log(df['Population'])
df['Density']  = np.log(df['Density'])
df['Area']  = np.log(df['Area'])
df['Polution']  = np.log(df['Polution'])
df.head()
```

```python
df.loc[:,'State':'County']
```

```python
#geneName = ['Population', 'Area','Density']
geneName = ['Area']
fig, ax = plt.subplots(1, 1, figsize=(10,10))
fig.subplots_adjust(left=0.1, right=0.9, top=0.8, bottom=0.2, wspace=0.5)

for idx,gene in enumerate(geneName):
    ax.set_title('Polution \nvs\n '+gene)
    ax.scatter(df[gene],df['Polution'], c='grey',marker='o' ,alpha=0.2, s = 50)
    ax.set_xlabel('Log('+gene+')')
    ax.set_ylabel('Log(Polution)')
    #ax.grid(True)

    X = sm.add_constant(df[gene])
    model = sm.OLS(df['Polution'], X)
    result = model.fit()
    print result.summary()
    ax.plot(df[gene], result.fittedvalues, 'k-', alpha=1,
            linewidth=3,label=('linear fitted\n'+r'$\beta:$'+'%.2f\n'%(result.params[1])+r'$R^2:$'+'%.2f'%(result.rsquared)))

    model_cubic = smf.ols(formula='Polution~ 1 + %s+ I(%s ** 2.0)'%(gene,gene), data=df)
    result = model_cubic.fit()
    print result.summary()
    ax.plot(df[gene], result.fittedvalues, 'g^', alpha=1,
            markersize=10,label=('cubic fitted\n'+r'$\beta_1:$'+'%.2f\n'%(result.params[1])+r'$\beta_2:$'+'%.2f\n'%(result.params[2])+r'$R^2:$'+'%.2f'%(result.rsquared)))
    ax.legend(loc='upper left', shadow=True, fontsize="small")

    #for i, txt in enumerate(df['Geo_name']):
    #  ax.annotate(txt, (df[gene].ix[i],df['Total_polution'].ix[i]))
    #  #ax.scatter(df['Total_polution'].ix[i],df[gene].ix[i], c=colorscale[i], s=5,  alpha=0.8)
plt.grid(True)
plt.show()
```

```python
filenamefinal = 'final_cleaned.csv'
df= pd.read_csv(filenamefinal, index_col = 0)
df.head()
```

```python
df['Total_population']  = np.log(df['Total_population'])
df['Population_density']  = np.log(df['Population_density'])
df['Area_total']  = np.log(df['Area_total'])

print df.head()
```

```python
df['Geo_name'].ix[1]
```

```python
geneName = ['Total_population', 'Population_density','Area_total']
fig, ax = plt.subplots(1, 3, figsize=(12,4))
fig.subplots_adjust(left=0.1, right=0.9, top=0.8, bottom=0.2, wspace=0.5)

for idx,gene in enumerate(geneName):
    ax[idx].set_title('Total_polution vs '+gene)
    ax[idx].scatter(df['Total_polution'], df[gene], c='b',marker='o',alpha=0.8)
    ax[idx].set_ylabel(gene)
    ax[idx].set_xlabel('Total_polution')
    ax[idx].set_xlim(-0.05,1.05)
    ax[idx].set_ylim(-0.05,1.05)
    ax[idx].grid(True)
    
    X = sm.add_constant(df['Total_polution'])
    model = sm.OLS(df[gene], X)
    result = model.fit()
    ax[idx].plot(df['Total_polution'], result.fittedvalues, 'g-', alpha=1,
                 linewidth=1.5,label=("linear fitted"))
    ax[idx].legend(loc='lower right', shadow=True, fontsize="small")

plt.show()
```
