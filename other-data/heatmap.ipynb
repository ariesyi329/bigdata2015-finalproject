```python
import argparse,csv,sys,os
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
import scipy as sc

filename = '../out_wed.csv'
df = pd.read_csv(filename)
df.head()
```

```python
df = df[['trip_distance', 'trip_time_in_secs','tip_amount','fare_amount','zipcode_pickup','zipcode_dropoff']]
df.head()
```

```python
df['tip_perc'] =(df['tip_amount']/df['fare_amount'])*100
df.head()
```

```python
df['zipcode_dropoff'] = df['zipcode_dropoff'].astype(str)
df['zipcode_pickup'] = df['zipcode_pickup'].astype(str)
df = df[df.zipcode_pickup != -1]
df = df[df.zipcode_dropoff != -1]
df = df[df['zipcode_pickup'].map(len) == 5]
df = df[df['zipcode_dropoff'].map(len) == 5]
```

```python
df.head()
```

```python
df['id'] = df.zipcode_pickup.map(str) +'|'+ df.zipcode_dropoff.map(str)
df.head()
```

```python
df.to_csv('out_minimized.csv')
```

```python
df['counter']=1
df.head()
```

```python
df_new = df.groupby('id').sum()
df_new['trip_distance'] =(df_new['trip_distance']/df_new['counter'])
df_new['trip_time_in_secs'] =(df_new['trip_time_in_secs']/df_new['counter'])
df_new['tip_amount'] =(df_new['tip_amount']/df_new['counter'])
df_new['fare_amount'] =(df_new['fare_amount']/df_new['counter'])
df_new['tip_perc'] =(df_new['tip_perc']/df_new['counter'])
df_new.head()



```

```python
df_new.to_csv('out_minimized-aggregated.csv')
```

```python
filename = 'out_minimized-aggregated.csv'
df_new = pd.read_csv(filename)
df_new.head()
```

```python
import scipy as sc

Index = list(set(df_new['pickup_zipcode'].tolist()))
Cols = list(set(df_new['dropoff_zipcode'].tolist()))
df_hm = pd.DataFrame(index=Index, columns=Cols)
df_hm = df_hm.fillna(0)
df_hm.head()
```

```python
for index, row in df_new.iterrows():
    pickzip = int(row['pickup_zipcode']) 
    dropzip = int(row['dropoff_zipcode'])
    df_hm.loc[pickzip,dropzip] = float(row['counter'])
df_hm.head()    
```

```python
df_hm.to_csv('out_minimized-heatmap-matrix.csv')
```

```python
from matplotlib import cm as cm

plt.figure(figsize=(30,30))
colors = [('white')] + [(cm.jet(i)) for i in xrange(1,256)]
plt.pcolor(df_hm, cmap=matplotlib.colors.LinearSegmentedColormap.from_list('new_map', colors, N=256))
plt.title('Trip numbers heatmap')
plt.yticks(np.arange(0, len(df_hm.index)), df_hm.index)
plt.xticks(np.arange(0, len(df_hm.columns)), df_hm.columns, rotation=90)
plt.colorbar()
plt.savefig('heatmap-tripnumber.png')
```
